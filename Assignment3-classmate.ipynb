In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
import matplotlib.pyplot as plt
import scipy.stats as st
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

%matplotlib notebook
```

```python
## Data arrays for the figure

means = np.mean(df, axis=1)
stds = np.std(df, axis=1)
yrs = df.index
stdErrs = stds/np.sqrt(df.shape[1])  ## standard sampling errors
errs95 = 1.96 * stdErrs  ## the 95% confidence interval for each year's ground truth mean
```

```python
## Data method for the figure

def getCDF(means, stdErrs, level):
    '''np.arrays for 1st 2 args, and a user-selected number for the 3rd'''
    diffs = level - means  ## diffs between selected level and sampled means
    devs = diffs / stdErrs  ## how many std devs from mu is the selected level
    pcts = [(1 - st.norm.cdf(d)) if d > 0 else st.norm.cdf(-d) for d in devs]  ## what % chance is mu higher than level
    return pcts
```

```python
## Build and display the figure

size = 10   # use this to tweak size of figure
n=42000  # initial level, chosen to show some variety in bar colors
cmap = plt.get_cmap(name='seismic', lut=20)  # color scheme for chart
barColors = [cmap(cdf) for cdf in getCDF(means, stdErrs, n)]  ## colors for bars based on initial n

plt.figure(figsize=(size, 0.8 * size))  # setting the aspect ratio
plt.xticks(yrs)
plt.ylim(0,70000)

ax1 = plt.gca()
for spine in ['top','right','bottom']:
    ax1.spines[spine].set_visible(False)
ax1.tick_params(bottom=False)

txt1 = '% chance that the true mean of the year sampled from is above selected level'
ax1.text(1991.8, 68000, txt1, fontsize=1.1*size)
txt2 = 'Click below to select a new level'
ax1.text(1992.6, 58000, txt2, fontsize=1.3*size)

# text.Text instance variable to update with mouse clicks
lev = ax1.text(1991.5, n, str(int(n)), color='w', bbox=dict(facecolor='green', alpha=0.7))

# draw the color key inset
ax2 = mpl_il.inset_axes(ax1, width='88%', height='8%', loc=9)
grad = np.linspace(0,1,101)
grad = np.stack((grad, grad))


def drawBars(barcolors):
    ax1.bar(yrs, means, yerr=errs95, capsize=2*size, width=1, ecolor='grey', edgecolor='k', color = barcolors)
    
def drawInset():
    ax2.imshow(grad, cmap=cmap)
    ax2.yaxis.set_tick_params(left=0)  # remove the gradient's ylabels
    ax2.set_yticklabels([])

def drawLevel(yval):
    ax1.axhline(y=yval, color='green', linewidth=2, label=str(yval), linestyle='dotted')
    lev.set_position((1991.5, yval))
    lev.set_text(str(int(yval)))
    
drawBars(barColors)
drawInset()
drawLevel(n)

## click responder, based on week 3 notebook:
def onclick(event):  
    for line in ax1.lines:
        if line.get_color() == 'green':
            ax1.lines.remove(line)
    yLevel = event.ydata
    barColors = [cmap(cdf) for cdf in getCDF(means, stdErrs, yLevel)] 
    drawBars(barColors)
    drawLevel(yLevel)
    

# tell mpl_connect we want to pass a 'button_press_event' into onclick when the event is detected
plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```
